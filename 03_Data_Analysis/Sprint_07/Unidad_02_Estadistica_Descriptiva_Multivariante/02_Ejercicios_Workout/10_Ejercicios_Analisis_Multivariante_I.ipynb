Usa scipy tal como lo hicimos en el workout para hacer el test de correlacin. Cul es el p-value?Qu podemos decir de la correlacin lineal entre "total" y "speeding"?

```python
from scipy.stats import pearsonr
pearsonr(df_car["total"], df_car["speeding"])
```

```python
corr_coef, p_value = pearsonr(df_car["speeding"], df_car["total"])
print(f"Correlacin: {corr_coef}")
print(f"P-value: {p_value}")
```

```python
# El pvalue es de 1.87e-06 significa que no es al azar. Hay una correlacin significativa positiva entre el exceso de velocidad y el total de accidentes 
# lo que nos dice que a mayor porcentaje de accidentes por exceso de velocidad, mayor el numero total de accidentes.
```
